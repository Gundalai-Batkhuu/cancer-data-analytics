```python
import pandas as pd
import matplotlib_inline as plt
import numpy as np
```

```python
#Importing data. Assigning column names and row indexes. 
df=pd.read_csv('cancer/data/old/data.csv',header=None,prefix='c')
#Selecting data category names and assigning them with simple indices.
columnNames=df.iloc[0:4] 
#Selecting data from the patient number 120 and onwards for analysis. 
df=df.iloc[23:]
df.set_index('c0',inplace=True)
```

```python
#Setting the number of columns that will be displayed to maximum. 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
columnNames
```

```python
#Creating a list of variables of interests.
variablesOfInterest=['Total ЗТЭ+ДТЭ EQD2','HRCTV 90 EQD2','volume','үе шат','хавдрын үр дүн','Химийн эмчилгээний давтамжийн тоо']
#Creating a dictionary object to store column names of variables of interest. 
variableName={}
#Updating variableName dictionary with variables of interest as items.
for item in variablesOfInterest:
    variableName.update({item:""})
#Selecting the column name of a variable of interest. Storing the column name in variableName. 
for (columnName, content) in columnNames.iteritems():
    for item in variablesOfInterest:
        if item in content.values:
            variableName[item]=columnName
#Manually searching for columns corresponding variables of interests. 
variableName['хавдрын үр дүн']='c200'
variableName['HRCTV 90 EQD2']='c54'
variableName['Химийн эмчилгээний давтамжийн тоо']='c140'
variableName['volume']='c40'
variableName['үе шат']='c12'
variableName
```

```python
#Removing columns and rows with all values missing
df.dropna(axis=1,how='all',inplace=True)
df.dropna(axis=0,how='all',inplace=True)
```

```python
## Part 7. Frequency of chemotherapy vs Cancer aftereffect
```

```python
#Data cleaning
#Dropping rows with null values
df['c157'].dropna(inplace=True)
#Transforming data into a list of strings
df['c157']=df['c157'].str.split().tolist()
```

```python
# Changing the phrases which state that cancer is decreased or disappeared into numerical proportions. The number 0 indicates the dissaperance of the cancer, while 1 indicates that no change of the cancer was observed. 
unusedWords=['Хавдар','хавдар']
for phrase in df['c157']:
    if 'арилсан' in phrase:
        phrase.remove('арилсан')
        phrase.insert(0,0)
    if 'багассан,' in phrase:
        phrase.remove('багассан,')
        if '90%' in phrase:
            phrase.remove('90%')
            phrase.insert(0,0.1)
        elif '80%' in phrase:
            phrase.remove('80%')
            phrase.insert(0,0.2)
        elif '80-90%' in phrase:
            phrase.remove('80-90%')
            phrase.insert(0,0.15) 
        elif '70-80%' in phrase:
            phrase.remove('70-80%')
            phrase.insert(0,0.25)
    if 'багассан' in phrase:
        phrase.remove('багассан')
        if '90%' in phrase:
            phrase.remove('90%')
            phrase.insert(0,0.1)
        elif '80%' in phrase:
            phrase.remove('80%')
            phrase.insert(0,0.2)
        elif '80-90%' in phrase:
            phrase.remove('80-90%')
            phrase.insert(0,0.15)
        elif '70-80%' in phrase:
            phrase.remove('70-80%')
            phrase.insert(0,0.25)      
    for word in unusedWords:    
        if word in phrase:
            phrase.remove(word)
```

```python
#Creating a new column representing the residual cancer proportion.
df['c200']=[phrase[0] for phrase in df['c157']]
```

```python
# The leftover phrases contain additional information on the cancer 
df['c157']=[phrase[1:] for phrase in df['c157']]
```

```python
#Creating a new dataframe consisting of variables of interest and uploading it into the repo.
df2=pd.DataFrame(df[['c136','c140','c200','c40','c12']])
df2.to_csv('AnalysisData.csv')
```
